# Mtodos hierrquicos de agrupamento - Tarefa

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler

import scipy.cluster.hierarchy as shc
```

## 1) Repetindo o que fizemos em aula

Primeiramente vamos repetir o que fizemos em aula com a base de pinguins.

1. Selecione as variveis quantitativas da base de pinguins
2. Elimine valores faltantes
3. Padronize as variveis para mdia zero e desvio padro 1
4. Faa um agrupamento hierrquico dividindo essa base em 3 grupos
5. Faa o Dendrograma

```python
# Carregando dataset

peng = sns.load_dataset('penguins')
peng.index.name='id'
peng.head()
```

```python
# Selecionando as variveis quantitativas e eliminando valores faltantes

peng_num = peng.select_dtypes('number').dropna()
```

```python
# Padronizando

peng_pad = StandardScaler().fit_transform(peng_num)
```

```python
# Definindo objeto de agrupamento com 3 clusters

clus_3 = AgglomerativeClustering(linkage="complete", 
                                 distance_threshold = None, 
                                 n_clusters=3)
```

```python
# Treinando o algoritmo

clus_3.fit(peng_pad)
```

```python
# Dendrograma

plt.figure(figsize=(15, 15))

plt.title("Penguins Dendrograms")

dend = shc.dendrogram(shc.linkage(peng_pad, method='complete'))
```
