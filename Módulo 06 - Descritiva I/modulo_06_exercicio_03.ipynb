```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from tabula import read_pdf

sinasc = pd.read_csv('sinasc_ro_2019.csv')
estrutura_sinasc = read_pdf('estrutura_sinasc.pdf')
```

#### 2) Categorize a variável APGAR5 conforme as seguintes quebras:

- 0 a 3: 'asfixia grave'
- 4 a 5: 'asfixia moderada'
- 6 a 7: 'asfixia leve'
- 8 a 10: 'normal'

```python
sinasc.loc[sinasc['APGAR5'] <= 10, 'categoria_apgar5'] = 'normal'
sinasc.loc[sinasc['APGAR5'] < 8, 'categoria_apgar5'] = 'asfixia leve'
sinasc.loc[sinasc['APGAR5'] < 6, 'categoria_apgar5'] = 'asfixia moderada'
sinasc.loc[sinasc['APGAR5'] < 4, 'categoria_apgar5'] = 'asfixia grave'

sinasc['categoria_apgar5'].value_counts()
```

#### 3) Monte um gráfico comparando a distribuição de APGAR5 por sexo para cada grupo de tempo de gestação (GESTACAO).

```python
ax = sns.boxplot(data=sinasc, y='APGAR5', x='SEXO', orient='v', hue='GESTACAO')
```

####  4) Monte uma tabela cruzada com APGAR5 nas colunas e GESTACAO nas linhas

```python
tabela_cruzada = pd.crosstab(sinasc['GESTACAO'], sinasc['APGAR5'])
tabela_cruzada
```

####  5) Da tabela cruzada do intem anterior, calcule os percentuais de linha. Ou seja, a distribuição da variável APGAR5 para cada classe de GESTACAO dado.

```python
percentual_linha_tabela_cruzada = tabela_cruzada.div(tabela_cruzada.sum(axis=1), axis=0)
percentual_linha_tabela_cruzada
```

####  6) Os médicos costumam classificar parto prematuro quando a gestação tem menos que 37 semanas. Faça um gráfico de barras da tabela cruzada acima e conclua se você acha que este conjunto de dados corrobora ou não com a afirmação dos médicos.

```python
percentual_linha_tabela_cruzada.plot.bar()

print('Concluo que o gráfico não corrobora com a afirmação dos médicos. Observamos que o APGAR5 é predominante nas faixas de 32 a 36 semanas e Menos de 22 semanas, por exemplo, o que está abaixo das 37 semanas. Além disso, nas demais faixas abaixo há predominância de um APGAR5 superior a 7.')
```
